```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
r=requests.get(url)
soup=BeautifulSoup(r.text,'html')
```

```python
table_tag=soup.find('table',class_="wikitable sortable")
table_tag.find_all('td')
```

```python
tr=table_tag.find_all('tr')
row_list=[]
for row in tr:
    td=row.find_all('td')
    d_list=[]
    for d in td:
        d_list.append(d.text)
    row_list.append(d_list)
rank=[x for x in range(1,51)]
row_list
```

```python
th=table_tag.find_all('th')      
th_list=[]
for i in th:
    th_list.append(i.text.strip())
col_name=th_list[:8]
col_name[6]='Headquarters'
col_name=col_name[1:]+['Rank']
col_name
```

```python
table=pd.DataFrame(row_list,columns=col_name)
table=table.dropna()
table['Rank']=rank
table=table[['Rank','Name','Industry','Revenue','Profit','Employees','Headquarters']]
table.head()
```

```python
for col_name in table:
    strip=[]
    for cell in table[col_name]:
        strip.append(str(cell).strip())
    table[col_name]=strip
table.head()
        
```

```python
IncDec=table_tag.find_all('img',alt=['Increase','Decrease'])
incdec_list=[]
for i in IncDec:
    incdec_list.append(i.get('alt'))
```

```python
rev_list=[]
count=0
for i in table['Revenue']:
    strip=i.strip('$').replace(',','')
    if incdec_list[count]=='Increase':
        rev_list.append(int(strip))
    else:
        rev_list.append(int('-'+strip))
    count+=1
```

```python
table['Revenue']=rev_list
table
```
