```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
ori_url='https://webscraper.io/test-sites/e-commerce/static/computers/laptops'
r=requests.get(ori_url)
soup=BeautifulSoup(r.text,'html')
```

```python
def scrape_all_href(url='https://webscraper.io/test-sites/e-commerce/static/computers/laptops',include=True):
    if include==True:
        href_list=[url]
    elif include==False:
        href_list=[]
    while url is not None:
        r=requests.get(url)
        soup=BeautifulSoup(r.text,'html')
        tag=soup.find('a',class_='page-link',rel='next')
        if not tag:
            break
        link=tag.get('href')
        complete_link='https://webscraper.io'+link
        href_list.append(complete_link)
        url=complete_link
    return href_list
```

```python
def scrape_one_page(link):
    r=requests.get(link)
    soup=BeautifulSoup(r.text,'html')
    all_item_tag=soup.find_all('div',class_='col-md-4 col-xl-4 col-lg-4')
    item_list=[]
    for item_tag in all_item_tag:
        name=item_tag.find('a',class_='title').get('title')
        price=item_tag.find('h4',class_="float-end price card-title pull-right").text
        description=item_tag.find('p',class_="description card-text").text
        rating=item_tag.find('p',{'data-rating':['1','2','3','4','5']}).get('data-rating')
        if rating=='1':
            rating=rating+' star'
        else: rating=rating+' stars'
        review=item_tag.find('p',class_='float-end review-count').string
        item_info=[name,description,price,rating,review]
        item_list.append(item_info)
    return item_list
```

```python
all_item=[]
for link in scrape_all_href():
    all_item+=scrape_one_page(link)
all_item
```

```python
table=pd.DataFrame(all_item,columns=['Product Name','Description','Price','Rating','Review'])
table
```
